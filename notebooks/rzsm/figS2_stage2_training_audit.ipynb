```python
import sys
from pathlib import Path
from IPython.display import Image, Markdown, display


def find_gptcast_root(start: Path) -> Path:
    p = start.resolve()
    for _ in range(8):
        if (p / 'gptcast').is_dir():
            return p
        if (p / 'GPTCast' / 'gptcast').is_dir():
            return (p / 'GPTCast').resolve()
        p = p.parent
    return start.resolve()


ROOT = find_gptcast_root(Path.cwd())
NOTEBOOK_DIR = ROOT / 'notebooks' / 'rzsm'
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
if str(NOTEBOOK_DIR) not in sys.path:
    sys.path.insert(0, str(NOTEBOOK_DIR))

from soilcast_stage2_paper_figures import plot_stage2_training_audit
```

```python
output_dir = ROOT / 'outputs' / 'stage2_forecaster_paper'

# 修改为你自己的训练 run 目录
run_map = {
    'baseline': ROOT / 'logs' / 'train' / 'runs' / 'REPLACE_ME_BASELINE_RUN',
    'physical_context': ROOT / 'logs' / 'train' / 'runs' / 'REPLACE_ME_STATIC_RUN',
}

audit_df, audit_summary = plot_stage2_training_audit(
    run_map=run_map,
    output_dir=output_dir,
    save_stem='figS2_stage2_training_audit',
)
print('已生成补充训练审计图:', audit_summary['png_path'])
display(Image(filename=audit_summary['png_path']))
display(Markdown('**建议英文 caption：**'))
display(Markdown(audit_summary['caption_en']))
display(Markdown('**建议中文 caption：**'))
display(Markdown(audit_summary['caption_zh']))
```
