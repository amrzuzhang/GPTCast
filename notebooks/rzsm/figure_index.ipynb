```python
import sys
from pathlib import Path


def find_gptcast_root(start: Path) -> Path:
    p = start.resolve()
    for _ in range(8):
        if (p / 'gptcast').is_dir():
            return p
        if (p / 'GPTCast' / 'gptcast').is_dir():
            return (p / 'GPTCast').resolve()
        p = p.parent
    return start.resolve()


ROOT = find_gptcast_root(Path.cwd())
print('当前 ROOT:', ROOT)
print('可用的第一阶段模型文件:')
for p in sorted((ROOT / 'models').glob('*.ckpt')):
    print('  -', p.name)

print('\n推荐直接打开的 notebook:')
for name in [
    'notebooks/rzsm/table1_stage1_summary.ipynb',
    'notebooks/rzsm/fig2_stage1_quantitative_comparison.ipynb',
    'notebooks/rzsm/fig3_stage1_case_studies.ipynb',
    'notebooks/rzsm/fig2_stage1_training_audit.ipynb',
    'notebooks/rzsm/table2_stage2_summary.ipynb',
    'notebooks/rzsm/fig4_stage2_leadwise_comparison.ipynb',
    'notebooks/rzsm/fig5_stage2_case_studies.ipynb',
    'notebooks/rzsm/figS2_stage2_training_audit.ipynb',
]:
    print('  -', ROOT / name)
```
