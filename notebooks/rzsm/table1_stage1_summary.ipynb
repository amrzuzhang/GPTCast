```python
import sys
from pathlib import Path
import pandas as pd


def find_gptcast_root(start: Path) -> Path:
    p = start.resolve()
    for _ in range(8):
        if (p / 'gptcast').is_dir():
            return p
        if (p / 'GPTCast' / 'gptcast').is_dir():
            return (p / 'GPTCast').resolve()
        p = p.parent
    return start.resolve()


ROOT = find_gptcast_root(Path.cwd())
NOTEBOOK_DIR = ROOT / "notebooks" / "rzsm"
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
if str(NOTEBOOK_DIR) not in sys.path:
    sys.path.insert(0, str(NOTEBOOK_DIR))

from soilcast_stage1_paper_figures import make_stage1_summary_table
```

```python
output_dir = ROOT / 'outputs' / 'stage1_tokenizer_paper'
benchmark_csv = output_dir / 'fig2_stage1_quantitative_comparison.csv'

if not benchmark_csv.exists():
    raise FileNotFoundError(
        f"Missing benchmark file: {benchmark_csv}. 这个 notebook 现在只负责生成 Table 1，不会自动重跑 Fig. 2。"
        f"请先运行 {ROOT / "notebooks/rzsm/fig2_stage1_quantitative_comparison.ipynb"} 生成 benchmark csv。"
    )

bench_df = pd.read_csv(benchmark_csv)
table_df, table_summary = make_stage1_summary_table(
    benchmark_df=bench_df,
    output_dir=output_dir,
    save_stem='table1_stage1_summary',
)
print('已生成表格文件:', table_summary['csv_path'])
table_df

print('\n建议英文 caption：')
print(table_summary['caption_en'])
print('\n建议中文 caption：')
print(table_summary['caption_zh'])
```
